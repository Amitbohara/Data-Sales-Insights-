Product and Variant: The separation allows us to manage variants (e.g., different colors or sizes) of a product. Each variant can be independently managed, including being discontinued without affecting the main product.

Price: Storing prices in a separate table allows tracking price changes over time, crucial for dynamic pricing.

Customer and Address: Keeping addresses separate from customers allows customers to have multiple addresses, and orders can reference a specific address.

Order and OrderItem: Separating orders and order items allows detailed tracking of each product variant ordered, maintaining detailed records for each order.

Relationships: Ensuring referential integrity through foreign keys maintains the consistency and reliability of the data.

```python
import random
import pandas as pd
from faker import Faker
from datetime import datetime, timedelta

# Initialize faker and random seed
fake = Faker()
random.seed(42)

# Generate Product Data
products = []
for i in range(1, 11):
    products.append({
        "ProductID": i,
        "Name": fake.word(),
        "Category": random.choice(["Clothing", "Groceries", "Electronics"]),
        "Description": fake.text(),
        "Status": random.choice(["active", "discontinued"])
    })

# Generate Variant Data
variants = []
variant_id = 1
for product in products:
    num_variants = random.randint(1, 3)  # Each product has 1 to 3 variants
    for _ in range(num_variants):
        variants.append({
            "VariantID": variant_id,
            "ProductID": product["ProductID"],
            "Name": f"{product['Name']} Variant {variant_id}",
            "Color": random.choice(["Red", "Green", "Blue"]),
            "Size": random.choice(["S", "M", "L", "XL"]),
            "Status": random.choice(["active", "discontinued"])
        })
        variant_id += 1

# Generate Price Data
prices = []
for variant in variants:
    num_prices = random.randint(1, 5)  # Each variant has 1 to 5 price changes
    current_date = datetime.now() - timedelta(days=365 * 2)  # 2 years ago
    for _ in range(num_prices):
        prices.append({
            "PriceID": len(prices) + 1,
            "VariantID": variant["VariantID"],
            "EffectiveDate": current_date.strftime("%Y-%m-%d"),
            "Price": round(random.uniform(10, 100), 2)
        })
        current_date += timedelta(days=random.randint(30, 180))  # Random date interval

# Generate Customer Data
customers = []
for i in range(1, 11):
    customers.append({
        "CustomerID": i,
        "FirstName": fake.first_name(),
        "LastName": fake.last_name(),
        "Email": fake.email(),
        "Phone": fake.phone_number(),
        "DateCreated": (datetime.now() - timedelta(days=random.randint(0, 365 * 2))).strftime("%Y-%m-%d")
    })

# Generate Address Data
addresses = []
for customer in customers:
    num_addresses = random.randint(1, 3)  # Each customer has 1 to 3 addresses
    for _ in range(num_addresses):
        addresses.append({
            "AddressID": len(addresses) + 1,
            "CustomerID": customer["CustomerID"],
            "Street": fake.street_address(),
            "City": fake.city(),
            "State": fake.state(),
            "Zip": fake.zipcode(),
            "Country": "USA",
            "IsPrimary": random.choice([True, False])
        })

# Generate Order Data
orders = []
order_items = []
order_id = 1
order_item_id = 1
for customer in customers:
    num_orders = random.randint(1, 5)  # Each customer places 1 to 5 orders
    for _ in range(num_orders):
        order_date = datetime.now() - timedelta(days=random.randint(0, 365 * 2))
        total_amount = 0
        num_items = random.randint(1, 5)  # Each order has 1 to 5 items
        for _ in range(num_items):
            variant = random.choice(variants)
            quantity = random.randint(1, 3)
            price = random.choice([p for p in prices if p["VariantID"] == variant["VariantID"]])["Price"]
            total_amount += price * quantity
            order_items.append({
                "OrderItemID": order_item_id,
                "OrderID": order_id,
                "VariantID": variant["VariantID"],
                "Quantity": quantity,
                "UnitPrice": price
            })
            order_item_id += 1
        orders.append({
            "OrderID": order_id,
            "CustomerID": customer["CustomerID"],
            "OrderDate": order_date.strftime("%Y-%m-%d"),
            "ShippingAddressID": random.choice([a for a in addresses if a["CustomerID"] == customer["CustomerID"]])["AddressID"],
            "TotalAmount": round(total_amount, 2)
        })
        order_id += 1

# Convert to DataFrames
df_products = pd.DataFrame(products)
df_variants = pd.DataFrame(variants)
df_prices = pd.DataFrame(prices)
df_customers = pd.DataFrame(customers)
df_addresses = pd.DataFrame(addresses)
df_orders = pd.DataFrame(orders)
df_order_items = pd.DataFrame(order_items)

# Display DataFrames
print("Products:")
print(df_products)
print("\nVariants:")
print(df_variants)
print("\nPrices:")
print(df_prices)
print("\nCustomers:")
print(df_customers)
print("\nAddresses:")
print(df_addresses)
print("\nOrders:")
print(df_orders)
print("\nOrderItems:")
print(df_order_items)
```

```python
import sqlite3

# Connect to SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('ecommerce.db')
cursor = conn.cursor()

# Create tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS Product (
    ProductID INTEGER PRIMARY KEY,
    Name TEXT,
    Category TEXT,
    Description TEXT,
    Status TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Variant (
    VariantID INTEGER PRIMARY KEY,
    ProductID INTEGER,
    Name TEXT,
    Color TEXT,
    Size TEXT,
    Status TEXT,
    FOREIGN KEY (ProductID) REFERENCES Product(ProductID)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Price (
    PriceID INTEGER PRIMARY KEY,
    VariantID INTEGER,
    EffectiveDate TEXT,
    Price REAL,
    FOREIGN KEY (VariantID) REFERENCES Variant(VariantID)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Customer (
    CustomerID INTEGER PRIMARY KEY,
    FirstName TEXT,
    LastName TEXT,
    Email TEXT,
    Phone TEXT,
    DateCreated TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Address (
    AddressID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    Street TEXT,
    City TEXT,
    State TEXT,
    Zip TEXT,
    Country TEXT,
    IsPrimary BOOLEAN,
    FOREIGN KEY (CustomerID) REFERENCES Customer(CustomerID)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS "Order" (
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    OrderDate TEXT,
    ShippingAddressID INTEGER,
    TotalAmount REAL,
    FOREIGN KEY (CustomerID) REFERENCES Customer(CustomerID),
    FOREIGN KEY (ShippingAddressID) REFERENCES Address(AddressID)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS OrderItem (
    OrderItemID INTEGER PRIMARY KEY,
    OrderID INTEGER,
    VariantID INTEGER,
    Quantity INTEGER,
    UnitPrice REAL,
    FOREIGN KEY (OrderID) REFERENCES "Order"(OrderID),
    FOREIGN KEY (VariantID) REFERENCES Variant(VariantID)
)
''')

conn.commit()
```

```python
import random
from faker import Faker
from datetime import datetime, timedelta

# Initialize faker and random seed
fake = Faker()
random.seed(42)

# Function to generate date strings
def random_date(start, end):
    return fake.date_between(start_date=start, end_date=end).strftime("%Y-%m-%d")

# Insert Product data
products = []
for i in range(1, 11):
    products.append((i, fake.word(), random.choice(["Clothing", "Groceries", "Electronics"]), fake.text(), random.choice(["active", "discontinued"])))

cursor.executemany('''
INSERT INTO Product (ProductID, Name, Category, Description, Status) VALUES (?, ?, ?, ?, ?)
''', products)

# Insert Variant data
variants = []
variant_id = 1
for product in products:
    num_variants = random.randint(1, 3)
    for _ in range(num_variants):
        variants.append((variant_id, product[0], f"{product[1]} Variant {variant_id}", random.choice(["Red", "Green", "Blue"]), random.choice(["S", "M", "L", "XL"]), random.choice(["active", "discontinued"])))
        variant_id += 1

cursor.executemany('''
INSERT INTO Variant (VariantID, ProductID, Name, Color, Size, Status) VALUES (?, ?, ?, ?, ?, ?)
''', variants)

# Insert Price data
prices = []
for variant in variants:
    num_prices = random.randint(1, 5)
    current_date = datetime.now() - timedelta(days=365 * 2)
    for _ in range(num_prices):
        prices.append((len(prices) + 1, variant[0], current_date.strftime("%Y-%m-%d"), round(random.uniform(10, 100), 2)))
        current_date += timedelta(days=random.randint(30, 180))

cursor.executemany('''
INSERT INTO Price (PriceID, VariantID, EffectiveDate, Price) VALUES (?, ?, ?, ?)
''', prices)

# Insert Customer data
customers = []
for i in range(1, 11):
    customers.append((i, fake.first_name(), fake.last_name(), fake.email(), fake.phone_number(), random_date("-2y", "today")))

cursor.executemany('''
INSERT INTO Customer (CustomerID, FirstName, LastName, Email, Phone, DateCreated) VALUES (?, ?, ?, ?, ?, ?)
''', customers)

# Insert Address data
addresses = []
for customer in customers:
    num_addresses = random.randint(1, 3)
    for _ in range(num_addresses):
        addresses.append((len(addresses) + 1, customer[0], fake.street_address(), fake.city(), fake.state(), fake.zipcode(), "USA", random.choice([True, False])))

cursor.executemany('''
INSERT INTO Address (AddressID, CustomerID, Street, City, State, Zip, Country, IsPrimary) VALUES (?, ?, ?, ?, ?, ?, ?, ?)
''', addresses)

# Insert Order and OrderItem data
orders = []
order_items = []
order_id = 1
order_item_id = 1
for customer in customers:
    num_orders = random.randint(1, 5)
    for _ in range(num_orders):
        order_date = random_date("-2y", "today")
        total_amount = 0
        num_items = random.randint(1, 5)
        for _ in range(num_items):
            variant = random.choice(variants)
            quantity = random.randint(1, 3)
            price = random.choice([p[3] for p in prices if p[1] == variant[0]])
            total_amount += price * quantity
            order_items.append((order_item_id, order_id, variant[0], quantity, price))
            order_item_id += 1
        orders.append((order_id, customer[0], order_date, random.choice([a[0] for a in addresses if a[1] == customer[0]]), round(total_amount, 2)))
        order_id += 1

cursor.executemany('''
INSERT INTO "Order" (OrderID, CustomerID, OrderDate, ShippingAddressID, TotalAmount) VALUES (?, ?, ?, ?, ?)
''', orders)

cursor.executemany('''
INSERT INTO OrderItem (OrderItemID, OrderID, VariantID, Quantity, UnitPrice) VALUES (?, ?, ?, ?, ?)
''', order_items)

conn.commit()
```

```python
# Run SQL queries
import pandas as pd

# Query 1: Top 5 customers with highest average order amounts in the last 6 months
query1 = '''
SELECT
    c.CustomerID,
    c.FirstName,
    c.LastName,
    AVG(o.TotalAmount) AS AvgOrderAmount
FROM
    "Order" o
JOIN
    Customer c ON o.CustomerID = c.CustomerID
WHERE
    o.OrderDate >= DATE('now', '-6 months')
GROUP BY
    c.CustomerID, c.FirstName, c.LastName
ORDER BY
    AvgOrderAmount DESC
LIMIT 5;
'''
result1 = pd.read_sql_query(query1, conn)
print("Query 1: Top 5 customers with highest average order amounts in the last 6 months")
print(result1)

# Query 2: Customers with lower order value this year compared to last year
query2 = '''
WITH OrderSums AS (
    SELECT
        CustomerID,
        strftime('%Y', OrderDate) AS OrderYear,
        SUM(TotalAmount) AS TotalOrderValue
    FROM
        "Order"
    GROUP BY
        CustomerID, OrderYear
)
SELECT
    o1.CustomerID,
    c.FirstName,
    c.LastName,
    o1.TotalOrderValue AS ThisYearOrderValue,
    o2.TotalOrderValue AS LastYearOrderValue
FROM
    OrderSums o1
JOIN
    OrderSums o2 ON o1.CustomerID = o2.CustomerID AND o1.OrderYear = strftime('%Y', 'now')
JOIN
    Customer c ON o1.CustomerID = c.CustomerID
WHERE
    o2.OrderYear = strftime('%Y', 'now', '-1 year')
    AND o1.TotalOrderValue < o2.TotalOrderValue;
'''
result2 = pd.read_sql_query(query2, conn)
print("\nQuery 2: Customers with lower order value this year compared to last year")
print(result2)

# Query 3: Cumulative purchases by a particular customer by product category
query3 = '''
SELECT
    c.CustomerID,
    c.FirstName,
    c.LastName,
    p.Category,
    SUM(oi.Quantity * oi.UnitPrice) AS CumulativePurchase
FROM
    "Order" o
JOIN
    OrderItem oi ON o.OrderID = oi.OrderID
JOIN
    Variant v ON oi.VariantID = v.VariantID
JOIN
    Product p ON v.ProductID = p.ProductID
JOIN
    Customer c ON o.CustomerID = c.CustomerID
WHERE
    c.CustomerID = 1  -- Replace with the specific CustomerID
GROUP BY
    c.CustomerID, c.FirstName, c.LastName, p.Category;
'''
result3 = pd.read_sql_query(query3, conn)
print("\nQuery 3: Cumulative purchases by a particular customer by product category")
print(result3)

# Query 4: Top 5 selling products with sales bifurcated by product variants
query4 = '''
WITH ProductSales AS (
    SELECT
        p.ProductID,
        p.Name AS ProductName,
        v.VariantID,
        v.Name AS VariantName,
        SUM(oi.Quantity) AS TotalQuantitySold,
        SUM(oi.Quantity * oi.UnitPrice) AS TotalSalesAmount
    FROM
        OrderItem oi
    JOIN
        Variant v ON oi.VariantID = v.VariantID
    JOIN
        Product p ON v.ProductID = p.ProductID
    GROUP BY
        p.ProductID, p.Name, v.VariantID, v.Name
),
TopProducts AS (
    SELECT
        ProductID,
        ProductName,
        SUM(TotalQuantitySold) AS TotalQuantitySold
    FROM
        ProductSales
    GROUP BY
        ProductID, ProductName
    ORDER BY
        TotalQuantitySold DESC
    LIMIT 5
)
SELECT
    tp.ProductID,
    tp.ProductName,
    ps.VariantID,
    ps.VariantName,
    ps.TotalQuantitySold,
    ps.TotalSalesAmount
FROM
    TopProducts tp
JOIN
    ProductSales ps ON tp.ProductID = ps.ProductID
ORDER BY
    tp.TotalQuantitySold DESC, ps.VariantID;
'''
result4 = pd.read_sql_query(query4, conn)
print("\nQuery 4: Top 5 selling products with sales bifurcated by product variants")
print(result4)
```
